```python
# using features incidenceRate, PctPublicCoverageAlone, povertyPercent, PctHS25_Over, PctPublicCoverage
# PctUnemployed16_Over, PctBachDeg25_Over, medIncome, PctEmployed16_Over, PctPrivateCoverage, PctPrivateCoverageAlone
# PctMarriedHouseholds, PctBachDeg18_24, PctEmpPrivCoverage, PctSomeCol18_24, PctHS18_24, MedianAge
# as these features have high correlation with deathrate
```

```python
data_corr = data_copy[['incidenceRate','PctPublicCoverageAlone','povertyPercent','PctHS25_Over','PctPublicCoverage','PctUnemployed16_Over','PctBachDeg25_Over','medIncome','PctPrivateCoverage','PctMarriedHouseholds','PctBachDeg18_24','PctEmpPrivCoverage','PctHS18_24','MedianAge']].copy()
```

```python
data_corr.head()
```

# Setting features and target

```python
target = data.deathrate
```

```python
features = data.drop('deathrate',axis = 1)
```

# Variance threshold selection

```python
from sklearn.feature_selection import VarianceThreshold
vt = VarianceThreshold(threshold =0.18)
vt.fit(features)
pd.DataFrame({'Variance': vt.variances_, 'select_status': vt.get_support()}, index=features.columns)
```

```python
data_vt = features.iloc[:,vt.get_support()]
```

```python
data_vt.head()
```

# Select K-Best method

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression 
from sklearn.model_selection import train_test_split
kbest = SelectKBest(score_func=f_regression, k=12)
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.25, random_state=42)
kbest.fit_transform(X_train, y_train)
data_selkBest = features.iloc[:,kbest.get_support()]
```

```python
data_selkBest.head()
```

# Different models 

```python
models =[]
r2=[]
rmse=[]
names=[]
fsel_list = []
ftrans_list = []
fscal_list = []
```

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
models.append(lr_model)
#names.append(lr_model.__class__.__name__)
```

```python
from sklearn.tree import DecisionTreeRegressor
tr_model = DecisionTreeRegressor(max_depth=3)
models.append(tr_model)
#names.append(tr_model.__class__.__name__)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gr_model = GradientBoostingRegressor()
models.append(gr_model)
#names.append(gr_model.__class__.__name__)
```

```python
from sklearn.ensemble import RandomForestRegressor
fr_model = RandomForestRegressor(n_estimators=100, max_depth=3)
models.append(fr_model)
#names.append(fr_model.__class__.__name__)
```

```python
from sklearn.ensemble import AdaBoostRegressor
ar_model = AdaBoostRegressor(n_estimators=100)
models.append(ar_model)
#names.append(ar_model.__class__.__name__)
```

```python
from sklearn.linear_model import SGDRegressor
sgd_model = SGDRegressor(max_iter=1000, tol=1e-3)
models.append(sgd_model)
#names.append(sgd_model.__class__.__name__)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import mean_squared_error
```

# Regression models with correlation based feature selection

```python
X_train, X_test,y_train, y_test = train_test_split(data_corr, target, test_size=0.25, random_state=42)
```

```python
for model in models:
    pipe = Pipeline([('regressor', model)])
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    r2.append(pipe.score(X_test, y_test))
    rmse.append(mean_squared_error(y_test, y_pred, squared=False))
    fsel_list.append("Correlation")
    ftrans_list.append("None")
    fscal_list.append("None")
    names.append(model.__class__.__name__)
```

# Regression models with Correlation based selected features and   Polynomial Features transformation

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree=2, include_bias=False, interaction_only=False)
fit_pf = pf.fit(data_corr)
poly = pd.DataFrame(pf.transform(data_corr), columns=fit_pf.get_feature_names(data_corr.columns))
```

```python
X_train, X_test,y_train, y_test = train_test_split(poly, target, test_size=0.25, random_state=42)
```

```python
for model in models:
    pipe = Pipeline([('regressor', model)])
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    r2.append(pipe.score(X_test, y_test))
    rmse.append(mean_squared_error(y_test, y_pred, squared=False))
    fsel_list.append("Correlation Based")
    ftrans_list.append("Poly 2")
    fscal_list.append("None")
    names.append(model.__class__.__name__)
```

# Regression models with Correlation based selected features and a feature scaling method

```python
X_train, X_test,y_train, y_test = train_test_split(data_corr, target, test_size=0.25, random_state=42)
for model in models:
    pipe = Pipeline([('scaler', RobustScaler()),('regressor', model)])
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    r2.append(pipe.score(X_test, y_test))
    rmse.append(mean_squared_error(y_test, y_pred, squared=False))
    fsel_list.append("Correlation Based")
    ftrans_list.append("None")
    fscal_list.append("Robust")
    names.append(model.__class__.__name__)
```
